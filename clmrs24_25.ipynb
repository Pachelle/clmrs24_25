```python
# Step 1: Create sets of child_ids from dfhhh and dfhhl
heavy_ids = set(dfhhh['child_id'])
light_work_ids = set(dfhhl['child_id'])
```

```python
# Step 2: Prepare a mapping of hours worked from dfhhl
# Ensure those columns are numeric
dfhhl['worklight_hours_schl'] = pd.to_numeric(dfhhl['worklight_hours_schl'], errors='coerce')
dfhhl['worklight_hours_we'] = pd.to_numeric(dfhhl['worklight_hours_we'], errors='coerce')
```

```python
# Set child_id as index for mapping
work_hours_map = dfhhl.set_index('child_id')[['worklight_hours_schl', 'worklight_hours_we']]
```

```python
work_hours_map
```

```python
# Step 3: Merge the hours info into dfhhc based on child_id (only brings in if exists)
dfhhc = dfhhc.merge(work_hours_map, how='left', left_on='child_id', right_index=True)
```

```python
# Step 4: Define the child labor rule
def is_child_labor(row):
    age = row['child_age']
    cid = row['child_id']
    hours_week = row.get('worklight_hours_schl', 0) or 0.0
    hours_weekend = row.get('worklight_hours_we', 0) or 0.0
    total_hours = hours_week + hours_weekend

    if (cid in heavy_ids) or (cid in light_work_ids and age < 13) or (age >= 13 and (hours_week > 2.0 or hours_weekend > 4.0 or total_hours >= 10.0)):
        return 'CL'
    else:
        return 'NoCL'
```

```python
# Step 5: Apply logic
dfhhc['child_labor'] = dfhhc.apply(is_child_labor, axis=1)
```

```python
# Step 1: Create a household-level child labor indicator from dfhhc
# We'll flag households where at least one child is in child labor

dfhhc['child_labor_flag'] = dfhhc['child_labor'] == 'CL'
```

```python
# Group by interview__id and check if any child is in labor
hh_child_labor = dfhhc.groupby('interview__id')['child_labor_flag'].any().reset_index()
```

```python
# Convert the boolean to 'Yes'/'No' or keep as boolean
hh_child_labor['child_labor_household'] = hh_child_labor['child_labor_flag'].map({True: 'Yes', False: 'No'})
```

```python
# Step 2: Merge into dfhh
dfhh = dfhh.merge(hh_child_labor[['interview__id', 'child_labor_household']], on='interview__id', how='left')
```

```python
# Optional: Fill NaN with 'No' if households without children should be considered as having no child labor
dfhh['child_labor_household'] = dfhh['child_labor_household'].fillna('No')
```
